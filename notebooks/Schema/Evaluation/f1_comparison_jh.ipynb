```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
all_results = pd.read_excel('../../../src/data/data/CSV_files/All_Results.xlsx')
all_results
```

```python
#add column to check whether F1 Score is larger for TURL compared to Tinybert, Roberta, Bert & Distilbert 
all_results['tinybert_vs_turl'] = np.where(all_results['Large_F1_Score']>=all_results['Tinybert'], True, False)
all_results['roberta_vs_turl'] = np.where(all_results['Large_F1_Score']>=all_results['Roberta'], True, False)
all_results['bert_vs_turl'] = np.where(all_results['Large_F1_Score']>=all_results['bert'], True, False)
all_results['distilbert_vs_turl'] = np.where(all_results['Large_F1_Score']>=all_results['distilbert'], True, False)
all_results
```

### Column Types where TURL performs same or better than all baseline models
##### for the large train set

```python
#only show column types where the F1 Score of TURL for the large train set is larger than all baselines
all_results[(all_results['tinybert_vs_turl']==True) & (all_results['bert_vs_turl']==True) & (all_results['roberta_vs_turl']==True) & (all_results['distilbert_vs_turl']==True)]
```

### Column Types where TURL performs worse than all baseline models
##### for the large train set

```python
#only show column types where the F1 Score of TURL for the large train set is smaller than all baselines
all_results[(all_results['tinybert_vs_turl']==False) & (all_results['bert_vs_turl']==False) & (all_results['roberta_vs_turl']==False) & (all_results['distilbert_vs_turl']==False)]
```

### Column Types where TURL performs better than Tinybert

```python
#only show column types where the F1 Score of TURL for the large train set is larger than all baselines
all_results[(all_results['tinybert_vs_turl']==True)]
```

### Column Types where TURL performs whorse than Tinybert

```python
#only show column types where the F1 Score of TURL for the large train set is larger than all baselines
all_results[(all_results['tinybert_vs_turl']==False)]
```
