```python
# find tables with column headers \\"byartist\\"
count = 0
with progressbar.ProgressBar(max_value=len(test_files)) as bar:
    inputdir = '../../../src/data/schemafiltereddata/TrainTestTables/Test/'
    files = [file for file in os.listdir(inputdir) if file.endswith('.json.gz')]
    df = pd.DataFrame ()
    find_file = []
    for file in files:
        df = pd.read_json(os.path.join(inputdir+file), compression = 'gzip', lines=True)
        #cat , rest = file.split('_', 1) #get category for file
        for col in df.columns:
            if '\\"byartist\\"' in col:
                find_file.append(file)
        count += 1
        bar.update(count)
    print(find_file)
```
