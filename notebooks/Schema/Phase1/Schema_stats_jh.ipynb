```python
import os
import pandas as pd
import json
import gzip
import fasttext
import numpy as np
```

```python
path = '/work-ceph/bizer-tp2021/data_integration_using_deep_learning/src/data/data/LocalBusiness/All'
```

```python
files = [file for file in os.listdir(path) if file.endswith('.json.gz')]
```

```python
column_count = []

for file in files:
    # read file
    df = pd.read_json(os.path.join(path, '{}'.format(file)), compression='gzip', lines=True)

    for x in df.columns:
        column_count.append(x)
```

```python
y_dict = {i:column_count.count(i) / len(files) for i in column_count}
display(y_dict)
```

```python
dict_sorted = dict(sorted(y_dict.items(), key=lambda item: item[1]))
```

```python
(pd.DataFrame.from_dict(data=dict_sorted, orient='index').to_csv('Person_stats.csv', header=False))
```
