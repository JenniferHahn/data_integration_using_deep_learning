```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
path = '../docs/'
```

```python
english_tables = pd.read_csv(path + r'EnglishDetection.csv', sep=';')
display(english_tables)
```

```python
plt = english_tables.plot(x='Category', y = 'Tables Left', kind='bar', figsize=(20,3))
plt.set_ylabel("Tables")
plt.set_xlabel("")
plt.legend(loc="upper right")
```

```python
final_tables = english_tables.iloc[1::2]
display(final_tables)
```

```python
plt = final_tables.plot(x='Category', y = 'Tables Left', kind='bar', figsize=(20,3))
plt.set_ylabel("# of Tables")
plt.set_xlabel("")
plt.legend(loc="upper right")

x = np.arange(len(final_tables['Tables Left']))
y = final_tables['Tables Left']
```

```python
final_sorted = final_tables.sort_values('Tables Left',ascending=False)
display(final_sorted)
final_tables = final_sorted
```

```python
plt = final_tables.plot(x='Category', y = 'Tables Left', kind='bar', figsize=(20,3))
plt.set_ylabel("# of Tables")
plt.set_xlabel("")
plt.legend(loc="upper right")

x = np.arange(len(final_tables['Tables Left']))
y = final_tables['Tables Left']

for i,j in zip(x, y):
    if (j < 5000): 
        plt.annotate(str(round(j, 2)), xy=(i,j))
```
