```python
zip_files = os.listdir(table_corpus_path)
```

```python
frames = []

count = 0
with progressbar.ProgressBar(max_value=len(zip_files)) as bar:
    for zip_file in zip_files:
        df = pd.read_json(table_corpus_path + '\{}'.format(zip_file), compression='gzip', lines=True)
        frames.append(df)
        count += 1
        if count % 100 == 0:
            cluster_df = pd.concat(frames, sort=False)
            print(cluster_df)
            break
        bar.update(count)
```
