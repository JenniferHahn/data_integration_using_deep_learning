```python
df['CombinedName'] = df['Class/Type'] + df['Properties'] + " - " + df['Match Class/Type'] + df['Match Property']
```

```python
df = df.iloc[:30].sort_values(by=['Tables w/ property', 'Matching Tables w/ property'], ascending=False)
```

```python
plotData = df
plot = plotData.plot(x='CombinedName', y=['Tables w/ property', 'Matching Tables w/ property'], kind='bar', figsize=(20,3))
plot.set_ylabel("Count tables")
plot.set_xlabel("")
plot.legend(loc="upper right")
```

```python
plotData = df
plot = plotData.plot(x='CombinedName', y=['Rows w/ property', 'Matching Rows w/ property'], kind='bar', figsize=(20,3))
plot.set_ylabel("Count rows in millions")
plot.set_xlabel("")
plot.legend(loc="upper right")
```
