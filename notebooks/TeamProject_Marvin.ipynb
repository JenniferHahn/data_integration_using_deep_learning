```python
import numpy as np
import pandas as pd
from os import walk
import json
import gzip
import matplotlib.pyplot as plt
```

```python
productPath = r'C:\Users\marvi\Desktop\SchemaOrgData\Product_top100' + '\\'
eventPath = r'C:\Users\marvi\Desktop\SchemaOrgData\Event_top100' + '\\'
jobPath = r'C:\Users\marvi\Desktop\SchemaOrgData\JobPosting_top100' + '\\'

productFilenames = next(walk(productPath), (None, None, []))[2]
eventFilenames = next(walk(eventPath), (None, None, []))[2]
jobFilenames = next(walk(jobPath), (None, None, []))[2]

productData = []
eventData = []
jobData = []

for product in productFilenames[:5]:
    with gzip.open(productPath + product, 'r') as dataFile:
        for line in dataFile:
            lineData = json.loads(line.decode('utf-8'))
            productData.append(lineData)
productDF = pd.DataFrame(productData)

for event in eventFilenames[:5]:
    with gzip.open(eventPath + event, 'r') as dataFile:
        for line in dataFile:
            lineData = json.loads(line.decode('utf-8'))
            eventData.append(lineData)
eventDF = pd.DataFrame(eventData)

for job in jobFilenames[:5]:
    with gzip.open(jobPath + job, 'r') as dataFile:
        for line in dataFile:
            lineData = json.loads(line.decode('utf-8'))
            jobData.append(lineData)
jobDF = pd.DataFrame(jobData)
```

```python
columnData = pd.read_csv(r'C:\Users\marvi\Desktop\ClassOverview.csv', sep=';').dropna().drop_duplicates()
# Value is NV -> not dropped by dropna
columnData = columnData.drop([84])
columnData = columnData.astype({'Matching Tables w/ property':'float64'})
columnData.head()
```

```python
columnData['CombinedName'] = columnData['Class/Type'] + columnData['Properties'] + " - " + columnData['Match Class/Type'] + columnData['Match Property']
```

```python
columnData = columnData.iloc[:30].sort_values(by=['Tables w/ property', 'Matching Tables w/ property'], ascending=False)
```

```python
columnData.head()
```

```python
filterValue = 300
plotData = columnData[(columnData['Tables w/ property'] > filterValue) | (columnData['Matching Tables w/ property'] > filterValue)]
plot = plotData.plot(x='CombinedName', y=['Tables w/ property', 'Matching Tables w/ property'], kind='bar', figsize=(20,3))
plot.set_ylabel("Tables in thousands")
plot.set_xlabel("")
plot.legend(loc="upper right")

```
