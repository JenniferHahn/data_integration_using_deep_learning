# This notebook was used to visualize the TURL training development

```python
import numpy as np
import os.path
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
visualization_path = '../../../../visualizations'
```

## Plot for results turl_transposed Product model

```python
# Step 1: default settings from TURL (5000-1500-5000 steps, batch size=20, 25 Epochen)
# Step 2: changed step sizes: 10-3-10 (25 Epochen)
# Step 3: Epochen erhöhen von 25 auf 50
# Step 4: learning rate von 5e-5 auf 5e-4
# Step 5: gradient acc. von 2 auf 1 (hat nichts gebracht)
# Step 6: gradient back to 2 + epsilon optimzer from 1e-8 to 1e-7
# Step 7: epsilon optimzer from 1e-7 to 1e-6
# Step 8: epsilon optimzer from 1e-6 to 1e-4
# Step 9: results all together: best result: 100 Epochs, batch size = 8, 5-1-5 steps, learning_rate = 5e-5, weight_decay = 0.01
# Step 10: best settings used on further preprocessed data ## still missing in overview
xs = [1,2,3,4,5,6,7,8,9]
ys = [0.0551, 0.6157, 0.7116, 0.7298, 0.7273, 0.7290, 0.7313, 0.7310, 0.7341]

plt.figure(figsize=(10,8))
plt.plot(xs,ys,'bo-')

for x,y in zip(xs,ys):
    label = "{:.4f}".format(y)
    plt.annotate(label, # this is the text
                 (x,y), # these are the coordinates to position the label
                 textcoords="offset points", # how to position the text
                 xytext=(0,8), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center
    
plt.yticks(np.arange(0, 0.8, step=0.05))
plt.ylabel('F1 micro score')
plt.xlabel('Settings')
plt.title('Different settings and their impact - Model TURL transposed for Product')
plt.savefig(os.path.join(visualization_path, 'TURL_transposed'))
plt.show()
```
