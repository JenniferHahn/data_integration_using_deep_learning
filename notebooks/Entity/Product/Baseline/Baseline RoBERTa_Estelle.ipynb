```python
import pandas as pd
import re
import numpy as np
import sklearn
import os
```

```python
file_path = '/content/drive/MyDrive/Colab/'
```

```python
df_train = pd.read_csv(os.path.join(file_path,'df_train'))
df_test = pd.read_csv(os.path.join(file_path,'df_test'))
```

```python
df_train
```

```python
train = df_train.rename(columns={"concat": "text","cluster_id_mapped":"label"})[['text','label']]
test = df_test.rename(columns={"concat": "text","cluster_id_mapped":"label"})[['text','label']]
```

```python
train
```

```python
num_columns = len(set.union(set(test.label.unique()), set(train.label.unique())))
```

```python
# arguments
train_args = {
    'num_train_epochs': 25,
    'overwrite_output_dir': True
}
```

```python
from simpletransformers.classification import ClassificationModel

# Create a ClassificationModel
model = ClassificationModel('roberta', 'roberta-base', num_labels=num_columns, args=train_args)
```

```python
model.train_model(train)
```

```python
result, model_outputs, wrong_predictions = model.eval_model(test)
```
