### Doc2Vec Modelling

```python
# convert into right format
df_final  = df_clean['new_column'].values
df_final[80000:80200]
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import nltk
#nltk.download('punkt')
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(df_final)]
```

```python
tagged_data[:100]
```

```python
# build model and vocabulary
model = Doc2Vec(vector_size=50, min_count = 5, epochs = 25, dm = 0) # distributed bow
model.build_vocab(tagged_data)
```

```python
# Train model
model.train(tagged_data, total_examples=model.corpus_count, epochs=25)
```

```python
# get top5 similar entities of tag 'xxx'
similar_doc = model.docvecs.most_similar('264183', topn = 50)
similar_doc
```

```python
index_list = []
for index, similarity in similar_doc:
    index = int(index)
    index_list.append(index)
df_clean.iloc[index_list]
```
