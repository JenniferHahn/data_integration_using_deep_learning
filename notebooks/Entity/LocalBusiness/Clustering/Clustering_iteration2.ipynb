```python
import numpy as np
import pandas as pd
import gzip
import json
```

```python
# path = r'../../../Results/'
# file = r'Concatenated_MFile.gz'
# pathFile = path + file
pathFile = 'Concatenated_MatchingFile_with_rest'
```

```python
dataList = []

with gzip.open(pathFile, 'r') as dataFile:
    for line in dataFile:
        lineData = json.loads(line.decode('utf-8'))
        dataList.append(lineData)
data = pd.DataFrame(dataList)
```

```python
print("number of entries:" , len(data))
print("number of unique tables:", data["origin"].nunique())
```

```python
clusters = data.groupby(['telephoneNorm']).size().reset_index(name='counts').sort_values('counts', ascending=False)
clusters = clusters.loc[clusters['counts'] > 50]

clusteredData = data[~data['telephoneNorm'].isin(clusters['telephoneNorm'])]
clusters = clusteredData.groupby(['telephoneNorm']).size().reset_index(name='counts').sort_values('counts', ascending=False)

clusteredData.loc[clusteredData['telephoneNorm'] == clusters.iloc[0]['telephoneNorm']]
```
